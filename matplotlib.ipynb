```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xp=np.array([3,4,5,6])
plt.plot(xp,marker='3')
plt.show()
```

```python
xp=np.array([3,4,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,marker='o')
plt.show()
```

```python
xp=np.array([3,4,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'o:r')
plt.show()
```

```python
xp=np.array([3,4,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'3--c') #-,.,-.,:
plt.show()
```

```python
xp=np.array([3,4,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'d--c',ms=2) #-,.,-.,:
plt.show()
```

```python
xp=np.array([3,4,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'3--r',ms=20,mec='b') #-,.,-.,:
plt.show()
```

```python
xp=np.array([3,2,5,2,4,7])
yp=np.array([2,5,3,4,6,8])
plt.plot(xp,yp,'3-c',mec='r',ms=30) #-,.,-.,:
plt.xlabel("hemant")
plt.ylabel("gami")
plt.show()
```

```python
xp=np.array([3,6,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'d-c',mec='r',mfc='r') #-,.,-.,:
plt.title("Hemant")
plt.show()
```

```python
xp=np.array([3,6,5,6])
yp=np.array([2,3,4,8])
plt.plot(xp,yp,'d-c',mec='r',mfc='r') #-,.,-.,:
font1 = {'family':'serif','color':'darkblue','size':20}
plt.title("Hemant",fontdict = font1,loc= 'Right')
plt.grid(color = 'green', linestyle = '--', linewidth = 1)
plt.show()
```
